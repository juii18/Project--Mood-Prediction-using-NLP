Step 2: Load model

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

Step 3: Test

```python
result = classifier("I love this movie")
print(result)
```

Step 4: Create test function

```python
def test(sentence):
    result = classifier(sentence)[0]

    if result["label"] == "POSITIVE":
        print("Positive 😊")
    else:
        print("Negative 😢")

    print("Confidence:", result["score"])
```

Step 5: Example

```python
test("This movie was amazing")
test("This was worst experience ever")
```
