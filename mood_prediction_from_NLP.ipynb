```python
!pip install datasets scikit-learn pandas

from datasets import load_dataset
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Load IMDB dataset
dataset = load_dataset("imdb")
train_data = dataset["train"]
test_data = dataset["test"]

# Convert to DataFrame
df_train = pd.DataFrame(train_data)
df_test = pd.DataFrame(test_data)

# Vectorize text
vectorizer = TfidfVectorizer(stop_words="english", max_features=10000)
X_train = vectorizer.fit_transform(df_train["text"])
X_test = vectorizer.transform(df_test["text"])

# Train model
model = MultinomialNB()
model.fit(X_train, df_train["label"])

print("Model Training Completed!")
```

```python
predictions = model.predict(X_test)
accuracy = accuracy_score(df_test["label"], predictions)
print("Model Accuracy:", accuracy)
```

```python
def test(sentence):
    vec = vectorizer.transform([sentence])
    pred = model.predict(vec)[0]
    if pred == 1:
        print("Positive ðŸ˜Š")
    else:
        print("Negative ðŸ˜¢")
```

```python
# Example tests
test("I feel so happy today")
test("This film was boring and terrible.")
test("I love my friends and family")
```
